## Data Description

The **Air Quality and Pollution Assessment** dataset, offers environmental metrics and demographic insights to facilitate the prediction of air quality. It encompasses data on various pollutants, meteorological conditions, and demographic factors, enabling comprehensive analysis of pollution levels and their potential health impacts. This dataset is valuable for developing machine learning models aimed at assessing and forecasting air quality across different regions.

**Input Variables:**
- Temperature (°C): Average temperature of the region.
- Humidity (%): Relative humidity recorded in the region.
- PM2.5 Concentration (µg/m³): Fine particulate matter levels.
- PM10 Concentration (µg/m³): Coarse particulate matter levels.
- NO2 Concentration (ppb): Nitrogen dioxide levels.
- SO2 Concentration (ppb): Sulfur dioxide levels.
- CO Concentration (ppm): Carbon monoxide levels.
- Proximity to Industrial Areas (km): Distance to the nearest industrial zone.
- Population Density (people/km²): Number of people per square kilometer in the region.


**Target Variable:**
`Air Quality level`:
- Good: Clean air with low pollution levels.
- Moderate: Acceptable air quality but with some pollutants present.
- Poor: Noticeable pollution that may cause health issues for sensitive groups.
- Hazardous: Highly polluted air posing serious health risks to the population.

## Feature Engineering

### Adding Binary Air Quality Column

```r
pollution$Air.Quality <- as.character(pollution$Air.Quality)
pollution$Air.Quality.New <- pollution$Air.Quality
pollution$Air.Quality.New[pollution$Air.Quality.New %in% c("Moderate", "Good")] <- "Acceptable"
pollution$Air.Quality.New[pollution$Air.Quality.New %in% c("Hazardous", "Poor")] <- "Unacceptable"

pollution$Air.Quality.New <- as.factor(pollution$Air.Quality.New)
head(pollution)
```

To conduct Logistic Regression, we created binary variable, 'Air.Quality.New', from our dependent variable 'Air Quality' This variable was either acceptable or unacceptable and was created by grouping the good and moderate classes and the poor and hazardous classes. This allowed us to run a logistic regressiona and look at the coefficients for the predictors. We also conducted step 

## Model Selection and Training

### Logistic Regression

#### Initial Model

```r
glm <-glm(Air.Quality.New~. - Air.Quality, data=traintransformed, family=binomial)
summary(glm)
```

```r
glm_probs <- predict(glm, testtransformed, type="response")
glm_pred <- ifelse(glm_probs>0.5,"Unacceptable", "Acceptable")
head(glm_pred)
```

```r
table(glm_pred, testtransformed$Air.Quality.New)
```

```r
# Ensure that both glm_pred and testtransformed$Air.Quality.New are factors with the same levels
glm_pred <- factor(glm_pred, levels = levels(testtransformed$Air.Quality.New))
testtransformed$Air.Quality.New <- factor(testtransformed$Air.Quality.New)

# Now compute the confusion matrix
glm_conf <- confusionMatrix(glm_pred, testtransformed$Air.Quality.New)
glm_conf
```

```r
mean(glm_pred!=testtransformed$Air.Quality.New) #error rate
mean(glm_pred==testtransformed$Air.Quality.New) #accuracy
```

**Assumption - Linearity**

```r
## Assumption 1 - Linearity of the Logit
library(dplyr)
library(tidyr)
library(ggplot2)

mydata <- testtransformed %>%
  dplyr::select_if(is.numeric) 
predictors <- colnames(mydata)
# Selects only numeric predictors from the test dataset (e.g., balance, income) for analysis.

# Bind the logit and tidying the data for plot
mydata <- mydata %>%
  mutate(logit = log(glm_probs/(1-glm_probs))) %>%
  pivot_longer(cols = -logit, names_to = "predictors", values_to = "predictor.value") #Reshapes the data into a tidy format, where each predictor is listed along with its values and the corresponding logit values.

ggplot(mydata, aes(logit, predictor.value))+
  geom_point(size = 0.5, alpha = 0.5) +
  geom_smooth(method = "loess") +
  theme_bw() +
  facet_wrap(~predictors, scales = "free_y")
```

For some predictors (`CO`, `Humidity`, `NO2`, `SO2`, `Temperature`), the line appears fairly linear, indicating the linearity assumption holds.

For others (`Proximity to Industrial Areas`, `PM2.5`, `population_densitiy`, `PM10`), the curve suggests a non-linear relationship, indicating the need for transformations or additional terms for these variables.

#### Final GLM model

```r
# GLM model with PM2.5 removed to reduce multicollinearity
final_glm <-glm(Air.Quality.New~. - Air.Quality - PM2.5, data=traintransformed, family=binomial)
summary(final_glm)
```

Based on our coefficients for our variables in our logistic regression, we see that CO concentrations and NO2 concentrations to have the highest magnitude of influence on the classification of air quality as acceptable or unnaceptable. PM10 was the least influential variable.

```r
final_glm_probs <- predict(final_glm, testtransformed, type="response")
final_glm_pred <- ifelse(final_glm_probs>0.5,"Unacceptable", "Acceptable")
head(final_glm_pred)
```

```r
table(final_glm_pred, testtransformed$Air.Quality.New)
```

```r
# Ensure that both glm_pred and testtransformed$Air.Quality.New are factors with the same levels
final_glm_pred <- factor(final_glm_pred, levels = levels(testtransformed$Air.Quality.New))
testtransformed$Air.Quality.New <- factor(testtransformed$Air.Quality.New)

# Now compute the confusion matrix
final_GLM <- confusionMatrix(final_glm_pred, testtransformed$Air.Quality.New)
final_GLM
```

```r
mean(final_glm_pred!=testtransformed$Air.Quality.New) #error rate
mean(final_glm_pred==testtransformed$Air.Quality.New) #accuracy
```

Our first GLM contained an error rate of .0305 and accuracy of .9695, so we actually see a decrease in our final GLM's accuracy due to removing variable PM2.5 becuase of multicollienearity.

```r
conf_matrix<- as.matrix(final_GLM$table)
desired_order <- c("Unacceptable", "Acceptable")

# Reorder the confusion matrix rows and columns to match the desired order
conf_matrix <- conf_matrix[desired_order, desired_order]
# Convert the confusion matrix to a data frame for plotting
conf_matrix_df <- as.data.frame(as.table(conf_matrix))
colnames(conf_matrix_df) <- c("Prediction", "Reference", "Count")
ggplot(conf_matrix_df, aes(x = Reference, y = Prediction, fill = Count)) +
  geom_tile() +
  geom_text(aes(label = Count), color = "black", size = 5) +
  scale_fill_gradient(low = "white", high = "purple") +
  labs(title = "GLM", x = "Reference", y = "Prediction") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

#### Linear Discriminant Analysis

```r
traintransformed$Air.Quality <- as.factor(traintransformed$Air.Quality)
testtransformed$Air.Quality <- factor(testtransformed$Air.Quality, levels = levels(traintransformed$Air.Quality))

lda_model <- lda(Air.Quality ~ . - Air.Quality.New - PM2.5, data = traintransformed)
lda_model
```

Looking at the models output we can see:
- **Prior Probabilities**: the distribution of of data points per class: Good 40%, Hazardous - 10.2%, Moderate ~30.1%, and Poor ~19.7%. 
- **Group Means**: shows the mean values for each variable and class, giving insight into the influence of variables on Air Quality. You can see that generally the Hazardous class has higher temperatures, humidity, PM10, NO2, SO2, and Population_Density than the Good Air Quality class. For Proximity to industrial area, its results are also pretty intuitive showing that larger proximity to industrial areas is Good Air quality and lower average proximity to industrial areas is hazardous.
- **Coefficiant Descriminents**: breaks down the 3 linear discriminents, explaining the coefficients behind each line. LD1 captures the majority of variation for distinguishing between air quality. We see that  NO2 and  CO have the greatest input for class variation of LD1. For LD2, which accounts for 1.86% of describining variation we see that Proximity to Industrial areas has the greatest influence. In our model LD3 only accounts for .02% of describing variation with Temperature and CO being its greatest influencers

```r
predictions <- predict(lda_model, testtransformed)
head(predictions)
```

```r
class(predictions$class)
LDA <- confusionMatrix(predictions$class,testtransformed$Air.Quality)
LDA
```

```r
# Load required library
library(ggplot2)

# Sensitivity data from confusion matrix
sensitivity_data <- data.frame(
  Class = c("Good", "Hazardous", "Moderate", "Poor"),
  Sensitivity = c(1.0000, 0.7990, 0.9396, 0.8268)
)


# Create the bar plot
ggplot(sensitivity_data, aes(x = Class, y = Sensitivity, fill = Class)) +
  geom_bar(stat = "identity", color = "black", width = 0.6) +
  geom_text(aes(label = sprintf("%.1f%%", Sensitivity * 100)), 
            vjust = -0.5, size = 4.5) +
  scale_y_continuous(labels = scales::percent_format(accuracy = 1), limits = c(0, 1)) +
  labs(
    title = "Per-Class Sensitivity of the LDA Model",
    x = "Class",
    y = "Sensitivity",
    fill = "Prediction Class"
  ) +
  theme_minimal(base_size = 14) +
  theme(legend.position = "none")
```

```r
conf_matrix<- as.matrix(LDA$table)
desired_order <- c("Hazardous", "Poor", "Moderate", "Good")

# Reorder the confusion matrix rows and columns to match the desired order
conf_matrix <- conf_matrix[desired_order, desired_order]
# Convert the confusion matrix to a data frame for plotting
conf_matrix_df <- as.data.frame(as.table(conf_matrix))
colnames(conf_matrix_df) <- c("Prediction", "Reference", "Count")
ggplot(conf_matrix_df, aes(x = Reference, y = Prediction, fill = Count)) +
  geom_tile() +
  geom_text(aes(label = Count), color = "black", size = 5) +
  scale_fill_gradient(low = "white", high = "purple") +
  labs(title = "LDA", x = "Reference", y = "Prediction") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
per_class_accuracy <- diag(conf_matrix) / rowSums(conf_matrix)

# Create a data frame for plotting
accuracy_df <- data.frame(
  Prediction = rownames(conf_matrix),
  Accuracy = per_class_accuracy
)

# Create the bar chart with percentage labels
ggplot(accuracy_df, aes(x = Prediction, y = Accuracy, fill = Prediction)) +
  geom_bar(stat = "identity", color = "black") +
  geom_text(aes(label = scales::percent(Accuracy, accuracy = 0.01)), vjust = -0.5, size = 4) + # Add labels above bars
  ylim(0, 1) + # Set the y-axis range from 0 to 1 (percentage scale)
  labs(title = "Per-Class Accuracy", x = "Prediction Class", y = "Accuracy Percentage") +
  scale_y_continuous(labels = scales::percent_format(accuracy = 1)) + # Format y-axis as percentage
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) + theme(panel.grid = element_blank())+  theme(legend.position = "none")

```

# Final Model Selection

Overall when comparing our models, it makes sense to compare the KNN and LDA against each other as they are both predicting air quality at the level of 4 classes. LDA had a better overall accuracy than KNN but the more noticeable difference was the sensitivities for the Hazardous class. LDA had a sensitivity of 0.7990 whereas KNN had a sensitivity of 0.7371. This is an important difference in reality because hazardous air quality days have the most signifigant potential health implications so being able to accurately predict those days becomes very important. So we decided to go with the LDA as our best predictive model. We ran the Logistic regression to be able to get a sense of which variables had the highest influence.
